Getting the initial stock data:

```python
import pandas as pd
import requests

api_key = "39IH1OIV8K9CGNHI"
ticker = "AAPL"
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol={ticker}&outputsize=full&apikey={api_key}"
response = requests.get(url)
data = response.json()["Time Series (Daily)"]

df = pd.DataFrame(columns=["TICKER", "DATE", "OPENING_PRICE", "CLOSING_PRICE", "% CHANGE"])

for date, prices in data.items():
    df = df.append({
        "TICKER": ticker,
        "DATE": pd.to_datetime(date, format="%Y-%m-%d"),
        "OPENING_PRICE": float(prices["1. open"]),
        "CLOSING_PRICE": float(prices["4. close"]),
        "% CHANGE": ((float(prices["4. close"]) - float(prices["1. open"])) / float(prices["1. open"])) * 100
    }, ignore_index=True)

df.to_csv(f"stock-data/{ticker}-historical-data.csv", index=False)
```

Convert the stock data into weekly, monthly, and yearly changes:

```python
import pandas as pd

stock_ticker = "TSLA"
df = pd.read_csv(f"stock-data/{stock_ticker}-historical-data.csv")

df["DATE"] = pd.to_datetime(df["DATE"])
df.set_index("DATE", inplace=True)

weekly_change = df.groupby("TICKER")["CLOSING_PRICE"].resample("W").ffill().pct_change()
monthly_change = df.groupby("TICKER")["CLOSING_PRICE"].resample("M").ffill().pct_change()
yearly_change = df.groupby("TICKER")["CLOSING_PRICE"].resample("Y").ffill().pct_change()

weekly_df = pd.DataFrame(weekly_change)
monthly_df = pd.DataFrame(monthly_change)
yearly_df = pd.DataFrame(yearly_change)

weekly_df.to_csv(f"stock-data-timeframes/{stock_ticker}-weekly_change.csv")
monthly_df.to_csv(f"stock-data-timeframes/{stock_ticker}-monthly_change.csv")
yearly_df.to_csv(f"stock-data-timeframes/{stock_ticker}-yearly_change.csv")

print("Weekly % Change:\n", weekly_df)
print("Monthly % Change:\n", monthly_df)
print("Yearly % Change:\n", yearly_df)
```

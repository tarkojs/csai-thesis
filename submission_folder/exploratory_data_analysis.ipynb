```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/stock_tweets_80k.csv')
value_counts = df['Stock Name'].value_counts()
filtered_counts = value_counts[value_counts >= value_counts.sum() * 0.05]

plt.pie(filtered_counts, labels=filtered_counts.index, autopct=lambda pct: f'{pct:.0f}%' if pct >= 5 else '')
plt.axis('equal')
plt.savefig('stock_dist.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# AAPL plot for final thesis version:

final_df = pd.read_csv('final_values_AAPL.csv')
fig, ax1 = plt.subplots()

final_df = final_df.iloc[:25]
color = 'tab:red'
ax1.plot(final_df['% CHANGE'], color=color)
ax1.set_xlabel('Day')
ax1.set_ylabel('% change', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(final_df['positive_percentage'], color=color)
ax2.set_ylabel('% positive polarity', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Daily stock price change plotted against avg. positive polarity')

plt.show()
```

```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data as pdr

ticker = "GOOGL"

start_date = "2020-01-01"
end_date = "2022-05-01"

msft = yf.Ticker("GOOGL")
print(msft.info)

yf.pdr_override()
data = pdr.get_data_yahoo(ticker, start=start_date, end=end_date)


df = pd.DataFrame(columns=["TICKER", "DATE", "OPENING_PRICE", "CLOSING_PRICE", "% CHANGE"])

df["TICKER"] = ticker
df["DATE"] = data.index
df["OPENING_PRICE"] = data["Open"]
df["CLOSING_PRICE"] = data["Close"]
df["% CHANGE"] = ((data["Close"] - data["Open"]) / data["Open"]) * 100

print(df.head(), df.tail())
```
